# **4.Data Visualization**

```python
# checking distribution of numerical column
sns.set_palette('muted')
from pylab import *
figure, axes = plt.subplots(figsize=(8,6))

subplot(2,3,1)
sns.distplot(df['price'])

subplot(2,3,2)
sns.distplot(df['minimum_nights'])

subplot(2,3,3)
sns.distplot(df['number_of_reviews'])

subplot(2,3,4)
sns.distplot(df['reviews_per_month'])

subplot(2,3,5)
sns.distplot(df['calculated_host_listings_count'])

subplot(2,3,6)
sns.distplot(df['availability_365'])

# avoid overlapping of plots
plt.tight_layout() 




```

- Most features are left Skewed

```python
# creating box plot
from pylab import *
figure, axes = plt.subplots(figsize=(8,6))

subplot(2,3,1)
sns.boxplot(y = df['price'])

subplot(2,3,2)
sns.boxplot(y = df['minimum_nights'])

subplot(2,3,3)
sns.boxplot(y = df['number_of_reviews'])

subplot(2,3,4)
sns.boxplot(y = df['reviews_per_month'])

subplot(2,3,5)
sns.boxplot(y = df['calculated_host_listings_count'])

subplot(2,3,6)
sns.boxplot(y = df['availability_365'])

# avoid overlapping of plots
plt.tight_layout() 



```

```python
# Total available Hotels Area wise
df.neighbourhood_group.value_counts()
```

```python
# Define the data
location = ['Manhattan', 'Brooklyn', 'Queens', 'Bronx', 'Staten Island']
values = [21660, 20095, 5666, 1090, 373]

# Create a bar graph using seaborn
sns.set_style('whitegrid')
sns.barplot(x=location, y=values)

# Add labels and titles
plt.xlabel('location')
plt.ylabel('Count')
plt.title('Properties per Neighbourhood Group',fontsize=14)

# Display the graph
plt.show()
```

- Manhattan and Brooklyn have most number of stays

```python
# most preferrd Room Type.
room=df.room_type.value_counts()
room.plot.pie(autopct='%1.1f%%',figsize=(8,7), fontsize=12)
plt.title('Properties Per Room Type' , fontsize=15)

```

- 52% guests prefer Entire home, 45.7% of guests prefer Private Room and only 2.4% of guests prefer Shared Room

```python
df.room_type.value_counts()
```

```python
# Define the data
room_type = ['Entire home/apt', 'Private room', 'Shared room']
value = [25407, 22319, 1158]

# Create a bar graph using seaborn
sns.set_style('whitegrid')
sns.barplot(x=room_type, y=value)

# Add labels and titles
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Room Type',fontsize=14)

# Display the graph
plt.show()
```
