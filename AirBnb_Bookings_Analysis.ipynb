# **3. Data Wrangling**
Making the dataset ready for analysis

**Dropping Null values** 

As there are lot of Null Values in the Dataset, So those Null Values have to be removed.

```python
# Dropping Null Values.
data_frame=df.dropna()
data_frame
```

```python
# Shape of Dataset after Dropping  Null Values.
data_frame.shape
```

```python
data_frame.head(10)
```

```python
data_frame.tail(5)
```

- **Creating New Column in the Dataframe**





```python
# New Column in the Dataset.
coordinates=data_frame.loc[:,['latitude','longitude']]
coordinates.head(5)
```

```python
data_frame['gps_coordinates'] = data_frame.apply(lambda row: (row['latitude'], row['longitude']), axis=1)
data_frame.head(5)
```

- **Creating New Dataframes as per the Availability Month wise in a year**

```python
# Dataframe of Hotels Available Through out the Year
full_year = data_frame[data_frame['availability_365']>364]
aval_all_year= full_year[['host_id','name','host_name','room_type','price','neighbourhood_group','neighbourhood_group','gps_coordinates','availability_365']]
aval_all_year.head(10)
```

```python
aval_all_year.shape
```

**Inference** - 

 841 Number of Hotels are Available throughout the year

```python
# Dataframe of Hotels Available for Six to Twelve months.
availability_more_than_six_months = data_frame[data_frame['availability_365']>183]
six_months_or_more=availability_more_than_six_months[['host_id','name','host_name','room_type','price','neighbourhood_group','neighbourhood_group','gps_coordinates','availability_365']]
six_months_or_more.head(5)
```

```python
six_months_or_more.shape
```

- This dataframe contains hotels which are available for 183 to 365 days in a year.

```python
# Dataframe of Hotels Available for one to Six months.
availability_less_than_six_months = data_frame[data_frame['availability_365']<183]
six_months_or_less=availability_less_than_six_months[['host_id','name','host_name','room_type','price','neighbourhood_group','neighbourhood_group','gps_coordinates','availability_365']]
six_months_or_less.head(10)
```

- This dataframe few hotels with zero availability through out the year, the hotels with zero availability have to be removed from this dataset.

```python
# Removing zero availability Stays
availability_less_than_six_months= six_months_or_less[six_months_or_less['availability_365']!=0]
availability_less_than_six_months.head(10)
```

- This dataframe shows the hotels which are available for between 1 to 183 days in a year

# **4.Data Visualization**

- Visualising the most preferred Room types using a pie chart

```python
# Pie chart of most preferrd Room Type.
room=data_frame.room_type.value_counts()
room.plot.pie(autopct='%1.1f%%',figsize=(8,7), fontsize=12)
plt.title('Preferred Room Type ' , fontsize=15)
```

- I have picked a pie chart because it expresses the whole relationship in a given data. In pie I believe that it's easy to explain the percentage comparison by the area covered in a circle and even we can use different color in our pie chart. So, I have used pie chart which helped me to get the percentage comparison more clearly and precisely.

- **Insights** - From the pie chart I have found out that Entire Home is mostly preferred by the guests over the other Room types. And the percentage count of Entire Home is 52.3,  percentage count of Private Room is 45.5 and the percentage of Shared Room is just 2.2. So I have got to know from the pie chart that many of the people have shown there interest in Entire Home.

- From the Visualization the Entire Room and Private Rooms are largely preferred by the Guests over the Shared Room; So the hotels should focus on the rooms with higher percentage demand inorder to boost their profits. i.e They should maintain more number of rooms of entire apt and private rooms than the shared rooms.
